```python
import requests

from bs4 import BeautifulSoup

def data_fetcher(url):
    dummy="https://www.sec.gov/Archives/edgar/data/35315/0000035315-03-000001.txt"
    r = requests.get(url)


    soup = BeautifulSoup(r.text)
    credit_default_tag=list()

    for tag in soup.find_all('tr'):
        for child in tag.childGenerator():
            try:
                list1=child.find_all('font')
                for i in list1 :
                    if (str(i.text) == "Credit Default Swap"):
                        credit_default_tag.append(i)
            except Exception as e:
                pass


    credit_default_final_tags=list()

    for k in credit_default_tag:
        try:
            if(len(k.text)>3):
                credit_default_final_tags.append(k)    

        except Exception as e:
            print e


    parent_tags=list()
    fields=list()

    for i in credit_default_final_tags:
        parent_tags.append(i.parent.parent.parent.parent)

    for each_tag in parent_tags:
        fields.append(each_tag.next_sibling.next_sibling)

    final_list =list()
    for j in fields:
        temp_list=list()
        for i in j.children :
            try:
                print i.p.font.get_text()
                temp_list.append(unicode(i.p.font.get_text()))

            except Exception as e:
                pass
        final_list.append(tuple(temp_list))

    return final_list
        

```
